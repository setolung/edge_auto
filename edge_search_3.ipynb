```python
import pyautogui
import random
import datetime
import time

# Word categories for random phrase generation
#nouns = ["cat", "dog", "bird", "tree", "house", "car", "book", "computer", "phone", "mountain", "river", "ocean", "sun", "moon", "star"]
#verbs = ["runs", "jumps", "sleeps", "eats", "flies", "swims", "reads", "writes", "sings", "dances", "thinks", "dreams", "explores", "creates", "invents"]
#adjectives = ["happy", "sad", "big", "small", "fast", "slow", "bright", "dark", "loud", "quiet", "soft", "hard", "hot", "cold", "shiny"]
#adverbs = ["quickly", "slowly", "loudly", "quietly", "carefully", "carelessly", "happily", "sadly", "brightly", "dimly", "softly", "roughly", "gently", "fiercely", "calmly"]
nouns = ["apple", "bridge", "cloud", "door", "elephant", "forest", "guitar", "hammer", "island", "jacket", "kettle", "lake", "mirror", "nest", "orange", "pencil", "quilt", "rain", "snake", "table", "umbrella", "village", "whale", "xylophone", "yarn", "zebra", "balloon", "cactus", "diamond", "engine", "feather", "glove", "horizon", "igloo", "lantern", "mushroom", "pyramid", "rocket", "saddle", "tunnel"]

verbs = ["bake", "climb", "dance", "erase", "float", "gather", "hike", "imagine", "jump", "kick", "listen", "melt", "navigate", "observe", "paint", "question", "repair", "sail", "tumble", "unwrap", "vanish", "whisper", "x-ray", "yell", "zip", "assemble", "bounce", "capture", "drift", "explore", "flick", "growl", "hover", "ignite", "juggle", "kneel", "leap", "mend", "nudge", "spin"]

adjectives = ["ancient", "bold", "calm", "dusty", "eager", "faint", "gentle", "harsh", "icy", "jolly", "keen", "lazy", "misty", "narrow", "odd", "plump", "quiet", "rugged", "sharp", "tidy", "uneven", "vivid", "warm", "zesty", "bitter", "crisp", "damp", "fluffy", "grim", "hollow", "juicy", "lively", "meek", "nifty", "pale", "rich", "sour", "tough", "vast", "wild"]

adverbs = ["abruptly", "bravely", "calmly", "daintily", "eagerly", "faintly", "gently", "harshly", "icily", "jovially", "keenly", "lazily", "mistily", "narrowly", "oddly", "plumply", "quietly", "roughly", "sharply", "tidily", "unevenly", "vividly", "warmly", "zestfully", "bitterly", "crisply", "damply", "fluffily", "grimly", "hollowly", "juicily", "livelily", "meekly", "niftily", "palely", "richly", "sourly", "toughly", "vastly", "wildly"]

def generate_random_phrase():
    structures = [
        "{adj} {noun} {verb} {adv}",
        "The {adj} {noun} {verb}",
        "{noun} {verb} {adv}",
        "{adv}, the {noun} {verb}",
        "{adj} {noun} {verb} the {noun}"
    ]
    
    structure = random.choice(structures)
    
    phrase = structure.format(
        adj=random.choice(adjectives),
        noun=random.choice(nouns),
        verb=random.choice(verbs),
        adv=random.choice(adverbs)
    )
    
    return phrase.capitalize()

def search_bing(repeat_count=29):
    # Coordinates for the Bing search field (you may need to adjust these)
    bing_search_field_x = 276  # Example X coordinate
    bing_search_field_y = 200  # Example Y coordinate
    
    for i in range(repeat_count):
        print(f"Iteration {i + 1} of {repeat_count}")  # Print the current iteration
        
        # Move the mouse to the Bing search field
        pyautogui.moveTo(bing_search_field_x, bing_search_field_y, duration=1)
        
        # Click to focus the search field
        pyautogui.click()
        
        # Clear the search field
        pyautogui.hotkey('ctrl', 'a')  # Select all text
        pyautogui.press('delete')      # Delete selected text
        time.sleep(0.5)                # Wait for 0.5 seconds
        
        # Generate a random phrase
        random_phrase = generate_random_phrase()
        
        # Type the random phrase into the search field
        pyautogui.write(random_phrase, interval=0.1)
        
        # Press Enter to perform the search
        pyautogui.press('enter')
        
        # Wait for 5 seconds before the next search
        time.sleep(5)

# Allow some time before executing the script
print("Script will start in 5 seconds. Please focus on your Bing search window.")
time.sleep(5)  # Wait for 5 seconds before running the function

# Run the search function
search_bing()

# Get current time
current_time = datetime.datetime.now()

# Print time in a readable format
print("Current time:", current_time.strftime("%H:%M:%S"))
# Calculate time 15 minutes later
time_later = current_time + datetime.timedelta(minutes=15)

# Print time 15 minutes later
print("15 minutes later:", time_later.strftime("%H:%M:%S"))
```
